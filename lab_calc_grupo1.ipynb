####Descrição de um conjunto

  Existem três recursos principais para descrever um conjunto e seus elementos:
  1. enumeração
  2. Propriedade Característica
  3. Diagrama de Venn


Como sabemos que matemática é uma escrita, cada recurso necessita ter a sua própria descrição, e quando queremos utilizar um recurso pela enumeração, quando citamos os elementos de um conjunto, devemos seguir a escrita dos elementos entre chaves.

</br>

*um exemplo rápido e próximo físicamente da gente*
</br>
dado um determinado conjunto de marca de celular dos alunos, temos:
                
        {samsung, iphone, xiaomi}

conjunto das cores das canetas que cada um trouxe para aula:
    
        {azul, preta, vermelha}

</br>

Esse foi um exemplo de conjuntos pequenos, poucos elementos.
Mas é interessante que essa forma de escrever também é útil quando queremos falar de um conjunto infinito.

conjunto de números ímpares positivos:

        {1, 3, 5, 7, 9, ...}

conjunto de números pares positivos:

        {2, 4, 6, 8, 10, ...}



conjunto dos múltiplos inteiros de 3:

      {0, 3, -3, 6, -6, ...}

</br>

E também é possível escrever com essa notação um intervalo quando o conjunto é finito

conjunto de números inteiros de 0 a 100:
    
    {0, 1, 2, ..., 100}
